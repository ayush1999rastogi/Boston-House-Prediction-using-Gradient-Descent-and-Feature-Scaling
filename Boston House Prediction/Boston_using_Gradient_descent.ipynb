```python
import numpy as np
import pandas as pd
from sklearn import preprocessing
```

# Loading the Training and Testing Datasets

```python
boston_train = np.genfromtxt("0000000000002417_training_boston_x_y_train.csv",delimiter = ",")
boston_test = np.loadtxt("0000000000002417_test_boston_x_test.csv",delimiter = ",")
print(boston_train.shape)
```

```python
df = pd.DataFrame(boston_train,columns = ['# CRIM', ' ZN', ' INDUS', ' CHAS', ' NOX', ' RM', ' AGE', ' DIS',
       ' RAD', ' TAX', ' PTRATIO', ' B', ' LSTAT', ' Y'])
df
```

# Feature Scaling the training data using StandardScaler

```python
Y_train = df[' Y']
del df[' Y']
```

```python
scaler = preprocessing.StandardScaler()
scaler.fit(df)
DF =pd.DataFrame(scaler.transform(df))
DF
```

# Adding all ones column in input 

```python
c = np.ones(boston_train.shape[0])
DF['C_'] = c
DF
```

```python
data_points = np.array(DF)
data_points
```

# Applying Gradient Descent Algorithm for Training

```python
def step_gradient(data_points, learning_rate, m):
    m_slope = np.zeros(DF.shape[1])
    M = len(data_points)
    for i in range(M):
        x = data_points[i]
        y = Y_train[i]
        for j in range(14):
            
            m_slope[j] += (-2/M) * ( (y - (m * x).sum())*x[j] )
            
    for j in range(14):
        m[j] = m[j] - (learning_rate * m_slope[j])
    
    return m
```

```python
def gd(data_points, learning_rate, num_iterations):
    m = np.zeros(DF.shape[1])
    for i in range(num_iterations):
        m = step_gradient(data_points, learning_rate, m)
        print(i, " Cost: ", cost(data_points, m))
    return m
```

```python
def cost(data_points, m):
    total_cost = 0
    M = len(data_points)
    for i in range(M):
        x = data_points[i]
        y = Y_train[i]
        total_cost += (1/M)*((y - (m*x).sum())**2)
    return total_cost
```

```python
def run():
    learning_rate = 0.08
    num_iterations = 600
    m = gd(data_points, learning_rate, num_iterations)
    print(m)
    return m
```

```python
m = run()
```
